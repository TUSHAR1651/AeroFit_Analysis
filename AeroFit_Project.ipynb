```python
frequency = [ 1, 31,  51, 55, 19, 4, 7, 9, 3]
IncomeIntervals = [(20000, 30000), (30000, 40000), (40000, 50000), (50000, 60000), (60000, 70000),
                    (70000, 80000), (80000, 90000), (90000, 100000), (100000, 110000)]

IncomeIntervals2 = [(50000,60000),(20000,30000), (30000,40000), (100000,110000), (40000,50000), (90000,100000),
               (80000,90000), (60000,70000), (70000,80000)]
frequency2 = [55 ,1, 31, 3, 51, 9, 7, 19 ,4]


plt.figure(figsize=(13, 6))
bars = plt.bar(range(len(IncomeIntervals)), frequency)
plt.xlabel('Income Intervals')
plt.ylabel('Frequency')
plt.title('Frequency within Income Intervals')
plt.xticks(range(0,9),[f"{i[0]}-{i[1]}" for i in IncomeIntervals])

for bar, count in zip(bars, frequency):
#     print(count)
#     print(bar.get_x)+bar.get_width()/2) 
    plt.text(bar.get_x()+bar.get_width()/2,bar.get_height(), count, ha='center')
plt.show()


plt.figure(figsize=(6, 7))
plt.pie(frequency2, labels=IncomeIntervals2, autopct='%1.1f%%')
plt.title('Pie Chart of Income Interval Frequencies')
plt.tight_layout()
plt.show()
```

## Observation from the above Graph

From the graph above, it is evident that individuals with a salary exceeding 30000 are more likely to purchase the product. However, the frequency of purchases decreases when the average salary exceeds 70000. Therefore, targeting individuals with salaries greater than 70000 may lead to an increase in sales, considering our product's higher price point. ```python

```python
Q1Amount = df['Income'].quantile(0.25)
Q3Amount = df['Income'].quantile(0.75)
IQR = Q3Amount - Q1Amount

lowerBoundAmount = Q1Amount - 1.5 * IQR
upperBoundAmount = Q3Amount + 1.5 * IQR


outliersAmount = df[(df['Income'] < lowerBoundAmount) | (df['Income'] > upperBoundAmount)]
m = np.median(df['Income'])
print("Median =",m)
print("Lower Bound = ", lowerBoundAmount)
print("Upper Bound = ", upperBoundAmount)

print("Outliers on the basis of Amount")
outliersAmount
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {'Product': ['KP281', 'KP481', 'KP781'],
        'Frequency': [80, 60, 40]}
dff = pd.DataFrame(data)

ProductAmounts = {
    'KP281': 1500,
    'KP481': 1750,
    'KP781': 2500,
}

dff['Amount Generated'] = dff['Frequency'] * dff['Product'].map(ProductAmounts)

plt.figure(figsize=(4, 5))
bars = plt.bar(dff['Product'], dff['Amount Generated'])

for bar, amount in zip(bars, dff['Amount Generated']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() ,amount, ha='center', va='bottom')

plt.xlabel('Product')
plt.ylabel('Amount Generated')
plt.title('Amount Generated by Products')
plt.show()


plt.figure(figsize=(4, 5))
plt.pie(dff['Amount Generated'], labels=dff['Product'], autopct='%1.1f%%', startangle=90)

plt.axis('equal') 
plt.title('Distribution of Amount Generated by Products')

plt.show()
```
